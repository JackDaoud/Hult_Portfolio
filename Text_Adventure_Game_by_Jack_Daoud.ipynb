`````python
##############################################################################
# Importing modules to be used
##############################################################################

import sys
from sys import exit
import random
import time
    
##############################################################################
# Constructing round_01 - Meet the Elven Oracle
##############################################################################

def round_01():
    
    """
This fantasy text adventure game consists of three rounds. The first is a brief 
dialogue with an elven oracle, the second is fortune telling using tarot cards 
to determine eligibility for a quest, and the third is the quest of a magic battle 
against an evil warlock who is keeping the queen of elves prisoner in a tower. 
You'll cast magic by rolling dice. 

Rounds:

- Round 01: Meet the Elven Oracle 

- Round 02: Tarot Card Reading

- Round 03: Save the Elven Queen



Known Bugs and/or Errors:

Sometimes the script reruns a scene (function) randomly... 
I've seen it happen twice to the function question_02() before round_02() is run.
I've also seen it happen to round_03() after the game was won via pass_or_fail().
This rerun bug happens only sometimes when I play the game, but it doesn't happen
to others who've played the game.
    """
    
    ##########################################################################
    # Globalizing variables that will be used throughout the code
    ##########################################################################
    
    global color_blue                                                         
    color_blue = '\033[34m'
    
    global color_end
    color_end = '\033[0m'   
    
    global color_red_underline
    color_red_underline = "\033[4;31m"
    
    global lines
    lines =\
"""-""" * 113
    
    input(prompt = """
<Press the Enter key to start the game>""")
    
    print(r"""
    
                  _______
         ..-'`       ````---.
       .'          ___ .'````.'SS'.
      /        ..-SS####'.  /SSHH##'.
     |       .'SSSHHHH##|/#/#HH#H####'.
    /      .'SSHHHHH####/||#/: \SHH#####\
   /      /SSHHHHH#####/!||;`___|SSHH###\
-..__    /SSSHHH######.         \SSSHH###\
`.'-.''--._SHHH#####.'           '.SH####/
  '. ``'-  '/SH####`/_             `|H##/
  | '.     /SSHH###|`'==.       .=='/\H|
  |   `'-.|SHHHH##/\__\/        /\//|~|/
  |    |S#|/HHH##/             |``  |
  |    \H' |H#.'`              \    |
  |        ''`|               -     /
  |          /H\          .----    /
  |         |H#/'.           `    /
  |          \| | '..            /
  |            /|    ''..______.'
   \          //\__    _..-. | 
    \         ||   ````     \ |_
     \    _.-|               \| |_
     _\_.-'   `'''''-.        |   `--.
 ''``    \            `''-;    \ /
          \      .-'|     ````.' -
          |    .'  `--'''''-.. |/
          |  .'               \|
          |.'        

    """)

    delay_print("""
Hello traveller... I've been expecting you!
    """)
    
    time.sleep(2)
    
    ##########################################################################
    # Defining function to ask user 1st question and collect input
    ##########################################################################
    #
    # This function will ask the user for an input and then run the next
    # function response_01() while calling their input to be carried over
    #
    ##########################################################################
    
    
    def question_01():
        user_input_01 = str(input(prompt = """
    Please reply by choosing an option using the number 1, 2, or 3:
       
        1) You have?
        2) I've been expecting you too...
        3) Who the heck are you? 
            
        > """))
    
        response_01(user_input_01)
    
    ##########################################################################
    # Defining function to give user a conditional response
    ##########################################################################
    #
    # This function tests the user input from question_01() using conditionals.
    #
    # There are three conditionals that will allow the user to procede.
    # 
    # Each conditionals consider the number in digit and text form as well as
    # unique words from each choice presented in question_01().
    # 
    # The else statement runs if the user's input is outside the conditionals 
    # then the function runs try_again() and calls response_01.
    #
    ##########################################################################
    
    def response_01(user_input_01):    
        if user_input_01 == "1"\
        or user_input_01.lower() == "one"\
        or "have" in user_input_01.lower():
            delay_print("""
Yes! I am the oracle Galadriel and so I have foreseen our encounter. In fact, I see and hear all things!

Before we go on, what are you called, traveller?           
           """)

        elif user_input_01 == "2"\
        or user_input_01.lower() == "two"\
        or "expect" in user_input_01\
        or "been" in user_input_01:
            delay_print("""
Wonderful! I'm the oracle Galadriel, nice to meet you... ?            
            """)

        elif user_input_01 == "3"\
        or user_input_01 == "three"\
        or "who" in user_input_01\
        or "heck" in user_input_01:
            delay_print("""
I am the renowned oracle Galadriel! I see and hear all things...

What are you called traveller? 
            """)

        else:
            try_again(response_01)
            
        question_02()
    
    ##########################################################################
    # Defining function to ask user a question and collect & validate input
    ##########################################################################
    #
    # This function starts off by running get_user_name() which is outlined 
    # at the end of this cell.
    # 
    # Afterwards, the fucntion asks the user for an input and runs the next
    # function response_02() while calling their input to be carried over
    #
    ##########################################################################
    
    def question_02():  
        
        get_user_name()
        
        delay_print(f"""
Greetings {color_blue + user_name + color_end}... Are you in search of a quest?\n""") 
        time.sleep(2)
        
        user_input_02 = str(input(prompt = """
    Please reply by choosing an option using the number 1, 2, or 3:   
        
        1) Depends... Where am I?
        2) Yes! I want to become a champion!
        3) Shouldn't you already know since you're an oracle...?
            
        > """))

        response_02(user_input_02)

    ##########################################################################
    # Defining function to give user a conditional response
    ##########################################################################
    #
    # This function tests the user input from question_02() using conditionals.
    #
    # There are three conditionals that will allow the user to procede.
    # 
    # Each conditionals consider the number in digit and text form as well as
    # unique words from each choice presented in question_02().
    # 
    # The else statement runs if the user's input is outside the conditionals 
    # then the function runs try_again() and calls response_02.
    #
    ##########################################################################
    
    
    def response_02(user_input_02):
        if user_input_02 == "1"\
        or user_input_02.lower() == "one"\
        or "depend" in user_input_02.lower()\
        or "where" in user_input_02.lower():
            delay_print(f"""
You're in Teldrassil {color_blue + user_name + color_end}, the land of night elves...

There is no time to tell you more! You must make haste! Our elven queen Elune is in grave danger...

Let us see whether or not you're meant to save her...
            """)
            
        elif user_input_02 == "2"\
        or user_input_02.lower() == "two"\
        or "yes" in user_input_02.lower()\
        or "become" in user_input_02.lower()\
        or "champion" in user_input_02.lower():
            delay_print(f"""
{color_blue + user_name + color_end}, not everyone is worthy of becoming a champion... 

But you're in luck! Our elven queen Elune is in grave danger, so you have a chance!

Let us see whether or not you're meant to save her...
            """)

        elif user_input_02 == "3"\
        or user_input_02.lower() == "three"\
        or "should" in user_input_02.lower()\
        or "already" in user_input_02.lower()\
        or "know" in user_input_02.lower()\
        or "since" in user_input_02.lower()\
        or "oracle" in user_input_02.lower():
            delay_print(f"""
Well {color_blue + user_name + color_end}, I know only what the Gods want me to know.

And I know that our elven queen Elune is in grave danger... But I don't know if you're fit for this quest

So come now, let's commune with the Gods and see whether or not you can proceed with this quest!
            """)
            
        else:
            try_again(response_02)
    
    question_01()
    
    round_02()

##############################################################################
# Constructing round_02 - Tarot Card Reading
##############################################################################

def round_02():

    ##########################################################################
    # Constructing dictionaries to loop through for tarot_card_reading
    ##########################################################################

    tarot_cards = {1 : "The Fool",
                   2 : "The Magician",
                   3 : "The High Priestess",
                   4 : "The Empress",
                   5 : "The Emperor",
                   6 : "The Hierophant",
                   7 : "The Lovers",
                   8 : "The Chariot",
                   9 : "Strength",
                   10 : "The Hermit",
                   11 : "Wheel of Fortune",
                   12 : "Justice",
                   13 : "The Hanged Man", 
                   14 : "Death", 
                   15 : "Temperance", 
                   16 : "The Devil", 
                   17 : "The Tower", 
                   18 : "The Star", 
                   19 : "The Moon",
                   20 : "The Sun",
                   21 : "Judgement", 
                   22 : "The World"}

    upright_tarot_description = {1 : "You have, are, or will embark on a new journey",
                                 2 : f"""{color_blue + user_name + color_end} is the embodiment of "where there's a will there's a way""",
                                 3 : "You're an intuitive and spiritual being", 
                                 4 : "You're in touch with your feminine side", 
                                 5 : "You're a strategic thinker who walks the walk", 
                                 6 : "You're playing it safe and following the status quo", 
                                 7 : "You have an extremely intimate relationship... \U0001F609", 
                                 8 : "You're in control and victorious!", 
                                 9 : "When things go wrong, youre as strong as a rock!", 
                                 10 : "You're a lonely wanderer following your inner light", 
                                 11 : "Good and bad are two sides of the same coin, and you know this!", 
                                 12 : f"Your decisions consider the long-term effects on you and others... {color_blue + user_name + color_end} the Goody two-shoes!", 
                                 13 : "You're 'suspending' certain actions in order to progress! - pun intended", 
                                 14 : f"{color_blue + user_name + color_end} is going to die and make way for {color_blue + user_name + color_end} 2.0", 
                                 15 : "You've mastered the Art of Not Giving a F*ck", 
                                 16 : "You've reached Aristotilean Eudaimonia and have tamed the beast within!", 
                                 17 : "Fundamental change is about to take place... Are you ready?", 
                                 18 : "Think of the birth of a star and all the planets that orbit it. Yep, that's you!", 
                                 19 : "The darkness of the night allows your intuition to take the driver's seat", 
                                 20 : f"{color_blue + user_name + color_end} is a luminous being bringing joy and happiness to those around you!", 
                                 21 : "You're able to complete a SWOT analysis on yourself!", 
                                 22 : "You're experiencing a sense of unity and wholeness... Did you eat a mushroom? \U0001F344\U0001F914"}

    reversed_tarot_description = {1 : "You're a fool! Stop yolo-ing and start planning",
                                  2 : "You're living an illusion! Open your eyes! \U0001F440",
                                  3 : "Stop thinking with your brain and start feeling with your gut!", 
                                  4 : "You can't help others unless you help yourself", 
                                  5 : "Becareful, there's a dictator in your life...", 
                                  6 : "You're feeling stuck in a cage of rules and structures", 
                                  7 : "There's inner or outer disharmony in your life...", 
                                  8 : "You're talking the talk but not walking the walk...", 
                                  9 : "Something's about to either turn you into Hulk or Courage the Cowardly Dog...", 
                                  10 : "You're going mad in your lonely search for truth, go make a friend!", 
                                  11 : "A dark cloud and black cat have been following you around...", 
                                  12 : "You're inconsiderate of the consequences of your actions on you and others... Smh...", 
                                  13 : "You feel like all your time and efforts are wasted", 
                                  14 : "You feel stuck in limbo as you resist letting go of the past..",
                                  15 : "You're stressed and anxious... You need to take a chill pill!", 
                                  16 : "You have an emptiness you can't satiate! You devil!", 
                                  17 : "A natural disaster is about to happen and no one is here to save you...", 
                                  18 : "You've lost faith and hope and what was once lit by the sun is now shrouded in darkness...", 
                                  19 : "You're on the dark side of the moon, don't jump off!", 
                                  20 : "You're not Always Look(ing) on the Bright Side of Life... Pun intended!", 
                                  21 : "You're too hard on yourself.. Don't forget that nobody's perfect!", 
                                  22 : "You look back at all your accomplishments and feel a strange emptiness... What's missing?"}
    
    ##########################################################################
    # Constructing dialogue of round_01
    ##########################################################################
    
    print(f"""
{lines}
    """)
    time.sleep(4)
    
    delay_print(f"""
Welcome to the Tarot Card Reading round {color_blue + user_name + color_end}!
    """)
    time.sleep(2)

    print(f"""
In order to determine your fate, we have to commune with the Gods using Tarot cards.

You will draw 3 random cards from a set of 22 Tarot cards. Each card has a 50/50 chance of being good or bad.

If 2 or more bad cards are drawn, you're unfit to save our queen.

But if 2 or more good cards are drawn, I'll teleport you to the queen's tower.

Let's see whether or not the odds are in your favor...
    """)
    time.sleep(4)
    
    ##########################################################################
    # Defining a function to shuffle cards (figuratively)
    ##########################################################################
    
    def draw_card(): 
        print(f"""
        
{lines}""")
        print(r"""
                                                                      _____
                                                                     |\ ~ /|
                                      You draw a card                |}}:{{|
                                                                     |}}:{{|
                                                                     |}}:{{| 
                                                                     |/_~_\|
        """)
        print(f"""
{lines}
        """)
        time.sleep(2)
    
    ##########################################################################
    # Defining a function to randomly draw 3 cards
    ##########################################################################
    #
    # This functions starts by instantiating two variables that will be used
    # for conditionals & looping.
    #
    # Note: the variable "fate" exists for the pass_or_fail() function in this
    # round.
    # 
    # The first loop is a while loop. While we have 20 or more cards, this
    # all code beneth the loop should keep running. This is to ensure the user 
    # gets 3 cards in total.
    #
    # Within that loop, subtract 1 card from the total count to ensure the
    # loop the doesn't run over and over and over and over and over and over.
    # (Pun intended)
    #
    # Afterwards, two placeholders - a list and an integer - are instantiated.
    # 
    # Within the while loop, a for loop is created to run through the first
    # dictionary - tarot_cards - in order to place card numbers in the 
    # card_numbers placeholder list
    #
    # Then random_card which starts at 0 changes based on a random choice from
    # the aforementioned placeholder list "card_numbers"
    #
    # Then conditionals are run to ensure that only 3 cards are drawn and
    # displayed for the user.
    #
    # Each conditional also .pop() (i.e. pull out from one place and place in
    # another) from tarot_cards dictionary to random_card variable. 
    # The popped item is a number denoting a tarot_card. 
    #
    # Each conditional also runs the function decide_fate() described below.
    # This is a conditional based function that will return 1 or -1 for "fate"
    # using the random_card that was previously popped.
    #
    # Finally pass_or_fail is run and passes "fate". This is to determine if
    # the user literally passes or fails this round. 
    #
    ##########################################################################
    
    def three_random_cards(tarot_cards):
        
        tarot_cards_count = 22
    
        fate = 0
    
        while tarot_cards_count >= 20:
            tarot_cards_count -= 1
            card_numbers = []
            random_card = 0

            for card in tarot_cards:
                card_numbers.append(card)
            random_card = random.choice(card_numbers)
            
            if tarot_cards_count == 21:
                input(prompt = """
                             
                             <Press the enter key to draw your first card>""")
                draw_card()
                delay_print("""
        1st card drawn: """ + color_blue + tarot_cards[random_card] + color_end)
                tarot_cards.pop(random_card)
                fate += decide_fate(random_card)
                time.sleep(2)
                
            elif tarot_cards_count == 20:
                input(prompt = """
                             
                             <Press the enter key to draw your second card>""")
                draw_card()
                delay_print("""
        2nd card drawn: """ + color_blue + tarot_cards[random_card] + color_end)
                tarot_cards.pop(random_card)
                fate += decide_fate(random_card)
                time.sleep(2)
                
            elif tarot_cards_count == 19:
                input(prompt = """
                             
                             <Press the enter key to draw your third card>""")
                draw_card()
                delay_print("""
        3rd card drawn: """ + color_blue + tarot_cards[random_card] + color_end)
                tarot_cards.pop(random_card)
                fate += decide_fate(random_card)
                time.sleep(2)

            else:
                break
                
        pass_or_fail(fate)
    
    ##########################################################################
    # Defining a function to decide fate (good or bad)
    ##########################################################################
    #
    # This function allows for two 50/50 possibilities to occur under the
    # variable "fate" which is used in the pass_or_fail() function
    #
    # If "fate" is good, the function returns 1 making the variable positive
    # 
    # If "fate" is bad, the function returns -1 making the variable negative
    #
    # This only works because we have 3 draws, so regardless of the outcome,
    # "fate" will either yield a positive or negative integer which is enough
    # to determine whether or not the user passes the round or fails the game.
    #
    ##########################################################################
    
    def decide_fate(random_card):
        good_fate = "\U0001F44D"
        bad_fate = "\U0001F44E"
        fate = [good_fate, bad_fate]
        random_fate = random.choice(fate)
        
        if random_fate == good_fate:
            delay_print(f"""
            
        {good_fate} Good fate: """ + upright_tarot_description[random_card])
            return 1
        
        if random_fate == bad_fate:
            delay_print(f"""
            
        {bad_fate} Bad fate: """ + reversed_tarot_description[random_card])
            return -1
        
        return 0
    
    ##########################################################################
    # Defining a function to determine whether or not user goes to round_03
    ##########################################################################
    #
    # This function runs a simple conditional. If "fate" is positive, the
    # user passes and moves on to round_03()
    #
    # If "fate" is negative, the user loses and the function fail() is run
    #
    ##########################################################################
    
    def pass_or_fail(fate):
        
        if fate > 0:
            delay_print("""

Congratulations! The Gods have spoken and you are fit to save the queen!
""")
            time.sleep(2)
            delay_print("""
Are you ready? I'll teleport you to the queen's tower in...
            """)
            time.sleep(1)
            print("""
                                                            3
            """)
            time.sleep(1)
            print("""
                                                            2            
            """)
            time.sleep(1)
            print("""
                                                            1            
            """)
            time.sleep(1)
            print(r"""
                                                       .-----.
                                             \ ' /   _/    )/
                                            - ( ) -('---''--)
                                             / . \((()\^_^/)()
                                              \\_\ (()_)-((()()
                                               '- \ )/\._./(()
                                                 '/\/( X   ) \
                                                 (___)|___/ ) \
                                                      |.#_|(___)
                                                     /\    \ ( (_
                                                     \/\/\/\) \\
                                                     | / \ |
                                                     |(   \|
                                                    _|_)__|_\_
                                                    )...()...(
                                                     | (   \ |     
                                                  .-'__,)  (  \
                                                            '\_-,
            """)
        else:
            print(f"""
            
Oops... You're unlucky today {color_blue + user_name + color_end}... That's 2 or more bad fate cards!
            """)
            fail()

    three_random_cards(tarot_cards)
    
    round_03()
    
##############################################################################
# Constructing round_03 - Save the Elven Queen
##############################################################################
    
def round_03(): 
    print(f"""
{lines}
    """)
    print(r"""
    
                                                |>>>
                                                |
                                            _  _|_  _
                           /`\             |;|_|;|_|;|
                                           \\.    .  /
            \,/                             \\:  .  /
                                             ||:   |
                                             ||:.  |
                                             ||:  .|
                                             ||:   |       \,/
                                             ||: , |            /`\
                                             ||:   |
                                             ||: . |
              __                            _||_   |
     ____--`~    '--~~__            __ ----~    ~`---,              ___
-~--~                   ~---__ ,--~'                  ~~----_____-~'   `~----~~     
    

    """)
    time.sleep(3)
    delay_print(f"""
Welcome to the final round {color_blue + user_name + color_end}!

    """)
    time.sleep(3)
    delay_print("""
The elven queen Elune is trapped in the tower and she needs your help!
    
    """)
    time.sleep(3)
    delay_print("""
The evil warlock Mathtrix is about to poison her...     

    """)
    print(r"""
                                                          .

                                                       .
                                             /^\     .
                                        /\   "V"
                                       /__\   I      O  o
                                      //..\\  I     .
                                      \].`[/  I
                                      /l\/j\  (]    .  O
                                     /. ~~ ,\/I          .
                                     \\L__j^\/I       o
                                      \/--v}  I     o   .
                                      |    |  I   _________
                                      |    |  I c(`       ')o
                                      |    l  I   \.     ,/
                                    _/j  L l\_!  _//^---^\\_  

    """)
    time.sleep(5)
    print("""
In order to save Elune, you'll have to cast a spell to counter Mathtrix's poisonous concoction!

I snuck 3 dice into your pocket. Roll them to cast your spell!

If the sum of all your rolls is an even number, you'll successfully counter Mathtrix's poison and save our queen!

But if the sum of all your rolls is an odd number, you will fail, and the elven queen will die... 

Are you ready?  
    """)
    time.sleep(2)

    ##########################################################################
    # Defining a function to roll dice (figuratively)
    ##########################################################################
    
    def roll_dice_figuratively(): 
        print(f"""
        
{lines}
                                                                   ________            
                                                                  /\       \           
                                                                 /()\   ()  \          
                                     You roll a dice            /    \_______\         
                                                                \    /()     /         
                                                                 \()/   ()  /          
                                                                  \/_____()/

{lines}
    """)
        time.sleep(2)
    
    ##########################################################################
    # Defining a function for whether or not the user wins or loses the game
    ##########################################################################
    #
    # This function is based on a simple conditional...
    #
    # sum_of_rolls variable is passed on from the roll_dice() function below.
    #
    # If the sum_of_rolls is divisible by 2 without having a remainder (i.e.
    # if it's an even number) the user wins.
    #
    # Otherwise, the user loses and the fail() function is run.
    #
    ##########################################################################
    
    def pass_or_fail(sum_of_rolls):
        
        if sum_of_rolls % 2 == 0:
            delay_print(f"""

Congratulations! Your rolls sum up to an even number and you've countered Mathtrix's poison!

You will forever be known as {color_blue + user_name + color_end} the champion who saved Elune, Teldrassil's elven queen!
            """)
            print(r"""
            
                                    _,._
                                   ,'   ,`-.
                        |.        /     |\  `.
                        \ \      (  ,-,-` ). `-._ __
                         \ \      \|\,'     `\  /'  `\
                          \ \      ` |, ,  /  \ \     \
                           \ \         `,_/`, /\,`-.__/`.
                            \ \            | ` /    /    `-._
                             \\\           `-/'    /         `-.
                              \\`/ _______,-/_   /'             \
                             ---'`|       |`  ),' `---.  ,       |
                              \..-`--..___|_,/          /       /
                                         |    |`,-,...,/      ,'     
                                         \    | |_|   /     ,' __  r-'',
                                          |___|/  |, /  __ /-''  `'`)  |  
                                       _,-'   ||__\ /,-' /     _,.--|  (
                                    .-'       )   `(_   / _,.-'  ,-' _,/
                                     `-------'       `--''       `'''
                                     
                                     
                        _________          _______    _______  _        ______  
                        \__   __/|\     /|(  ____ \  (  ____ \( (    /|(  __  \ 
                           ) (   | )   ( || (    \/  | (    \/|  \  ( || (  \  )
                           | |   | (___) || (__      | (__    |   \ | || |   ) |
                           | |   |  ___  ||  __)     |  __)   | (\ \) || |   | |
                           | |   | (   ) || (        | (      | | \   || |   ) |
                           | |   | )   ( || (____/\  | (____/\| )  \  || (__/  )
                           )_(   |/     \|(_______/  (_______/|/    )_)(______/ 
                                                                                  
                                     
            """)
            
            
        else:
            print(f"""
            
Oops... You're unlucky today {color_blue + user_name + color_end}... That's an odd number!
            """)
            fail()      
        
    ##########################################################################
    # Defining a function to roll dice and extract a random value
    ##########################################################################
    #
    # This function begins by instantiating 2 variables and 2 place_holders
    #   "dice" is because we only want to have 3 rolls in total
    #   "dice_numbers" is the list we'll derive our random choice from
    #   "rolls" is a place_holder to keep track of all 3 rolls the user does
    #   "sum_of_rolls" is for this round's pass_or_fail() function
    #
    # The function runs a loop while there's more than 0 dice.
    #
    # Within the loop, there are 3 conditionals to control all three rolls.
    #
    # Each conditional runs roll_dice_figuratively and extracts a random
    # choice from the list of dice_numbers. It also adds the roll to 
    # sum_of_rolls for this round's pass_or_fail() function which is finally
    # run at the end of the function while passing sum_of_rolls.
    #
    ##########################################################################
    
    def roll_dice():
        
        dice = 3
        dice_numbers = [1, 2, 3, 4, 5, 6]
        rolls = []
        sum_of_rolls = 0
       
        while dice > 0:
    
            if dice == 3:
                input(prompt = """ 
                            <Press the enter key to roll your first dice>""")
                roll_dice_figuratively()
                str(rolls.append(random.choice(dice_numbers)))
                print(f"""
1st roll: """ + color_blue + str(rolls[0]) + color_end)
                time.sleep(2)
                sum_of_rolls += rolls[0]
                
            elif dice == 2:
                input(prompt = """
                            <Press the enter key to roll your second dice>""")
                roll_dice_figuratively()
                str(rolls.append(random.choice(dice_numbers)))
                print(f"""
2nd roll: """ + color_blue + str(rolls[1]) + color_end)
                time.sleep(2)
                sum_of_rolls += rolls[1]
                
            elif dice == 1:
                input(prompt = """
                             <Press the enter key to roll your third dice>""")
                roll_dice_figuratively()
                str(rolls.append(random.choice(dice_numbers)))
                print(f"""
3rd roll: """ + color_blue + str(rolls[2]) + color_end)
                time.sleep(2)
                sum_of_rolls += rolls[2]
                input(prompt = """
                             <Press the enter key to to sum your roles>""")
            
            dice -= 1
            
        delay_print(f"""
Sum of rolls: """ + color_blue + str(sum_of_rolls) + color_end)
        time.sleep(4)
            
        pass_or_fail(sum_of_rolls)
        
    roll_dice()

##############################################################################
# Defining function for user_name input to be used globally
##############################################################################
    
def get_user_name():
    user_input = str(input(prompt = """
    Please input your name: """)).capitalize()
       
    global user_name     
    user_name = user_input
    

##############################################################################
# Defining function to delay print to imitate a typewriter-like style vibe
##############################################################################

def delay_print(text_place_holder):
    for character in text_place_holder:
        sys.stdout.write(character)
        sys.stdout.flush()
        time.sleep(0.03)

##############################################################################
# Defining recursive function for error handling through out the game
##############################################################################
#
# This is a recursive function that uses a function placeholder.
#
# Its purpose is to try the user input under certain functions (such as 
# question_02 for example) and see whether or not their input passes the
# conditionals (such as response_02). If it does, they move on in the game.
#
# If the user input doesn't pass the conditionals of the relevant function,
# the function itself is run again, and again, and again, and again.
#
# It's a loop function without the loop, and I think it's unbreakable!
#
##############################################################################
    
def try_again(function_place_holder):
    
    error_message_01 = """Error: That's not 1, 2 , or 3! Please try again..."""
    
    error_message_02 = """You can use "1" or "one" or a word unique to the sentence you want to choose!"""
    
    
    try:
        print(f"""
        {color_red_underline + error_message_01 + color_end}
        """)
        print(f"""
        {color_red_underline + error_message_02 + color_end}  
        """)
        
        user_input = str(input(prompt = """
        > """))
        
        function_place_holder(user_input)
        
    except ValueError:
        try_again(function_place_holder)

##############################################################################
# A function to allow the user the choice of which round to play after failing
##############################################################################
#
# Each conditionals consider the round number in digit and text form as well 
# as unique words from each choice presented.
# 
# The else statement runs if the user's input is outside the conditionals 
# then the function runs try_again() and calls which_round() - this function.
# i.e. the function loops under the else condition.
#
##############################################################################

def which_round():
    which_round_input = input(prompt = """
    Which round would you like to start at?
    
        1) Round 01 - Elven Oracle
        2) Round 02 - Tarot Card Reading
        3) Round 03 - Save the Princess
        
        > """)
        
    if "1" in which_round_input\
    or "one" in which_round_input.lower()\
    or "elven" in which_round_input.lower()\
    or "oracle" in which_round_input.lower():
        round_01()
            
    elif "2" in which_round_input\
    or "two" in which_round_input.lower()\
    or "tarot" in which_round_input.lower()\
    or "card" in which_round_input.lower()\
    or "reading" in which_round_input.lower():
        round_02()
            
    elif "3" in which_round_input\
    or "three" in which_round_input.lower()\
    or "save" in which_round_input.lower()\
    or "the" in which_round_input.lower()\
    or "princess" in which_round_input.lower()\
    or "dice" in which_round_input.lower():
        round_03()
            
    else:
        try_again(which_round)

##############################################################################
# Constructing fail function
##############################################################################

def fail():
    time.sleep(4)
    print(f"""

{'*' * 113}

    """)
    
    print(r"""
                                                 .-''''''-.
                                               .'          '.
                                              /   O      O   \
                                             :                :
                                             |                |
                                             :    .------.    :
                                              \  '        '  /
                                               '.          .'
                                                 '-......-'                     
                                             
                 _______  _______  _______  _______    _______           _______  _______ 
                (  ____ \(  ___  )(       )(  ____ \  (  ___  )|\     /|(  ____ \(  ____ )
                | (    \/| (   ) || () () || (    \/  | (   ) || )   ( || (    \/| (    )|
                | |      | (___) || || || || (__      | |   | || |   | || (__    | (____)|
                | | ____ |  ___  || |(_)| ||  __)     | |   | |( (   ) )|  __)   |     __)
                | | \_  )| (   ) || |   | || (        | |   | | \ \_/ / | (      | (\ (   
                | (___) || )   ( || )   ( || (____/\  | (___) |  \   /  | (____/\| ) \ \__
                (_______)|/     \||/     \|(_______/  (_______)   \_/   (_______/|/   \__/
                                                                            

    """)
    time.sleep(2)
    delay_print(f"""
Oh, I'm sorry {color_blue + user_name + color_end}, looks like you've lost!

Would you like to try your luck and play again? 

""")
    time.sleep(2)
    replay = str(input(f"""
    Please reply by choosing an option using the number 1, 2, or 3:
       
        1) Yes! This game was so much!
        2) No! This game wasn't fun...
        3) Maybe later!
            
        > """))  
    
    if replay == '1'\
    or replay.lower() == "yes"\
    or "y" in replay.lower():
        which_round()
        
    elif replay == '2'\
    or replay.lower() == "no"\
    or "n" in replay.lower():
        delay_print(f"Thanks for playing {user_name}")
        sys.exit(0)
    
    elif replay == '3'\
    or replay.lower() == "maybe":
        delay_print("Alrighty, hope to see you again soon!")
        sys.exit(0)
        
    else:
        try_again(fail)
        
##############################################################################
# Disclaimer: all ASCII art was borrowed from https://www.asciiart.eu
###############################################################################
        
round_01()
`````
